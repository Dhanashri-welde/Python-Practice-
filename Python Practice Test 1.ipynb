# 6.	Take Churned as a list
* a.Upper Case
* b.Replace = Stayed = not churned 


```python
Data = {'Customer Status':['Churned','Stayed','Churned','Stayed']}
df = pd.DataFrame(Data)
df

# upper
Churned_list = df[df['Customer Status'] == 'Churned']['Customer Status'].tolist()
Churned_list_upper = [Status.upper() for Status in Churned_list]

# Replace
df['Customer Status'] = df['Customer Status'].replace('Stayed','Not Churned')

print("All Churned Customer in upper case:",Churned_list_upper)
print("updated DataFrame:",df)
```
